```python
import pandas as pd
import numpy as np

df = pd.read_csv('historical-weather.csv',header=0,usecols=['air_pressure','air_temp','avg_wind_direction'])
columns = df.columns
columns
features = df
features
```

##### Preprocessing

Recieved from https://www.geeksforgeeks.org/how-to-scale-pandas-dataframe-columns/

```python
df_norm = (df-df.min())/(df.max()-df.min())
df_norm = df_norm.dropna()
df_norm
```

##### Part 1

```python
import matplotlib.pyplot as plt
import numpy as np
import random

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
count = 0
#colors = ['red','blue','green']

#how would I create more dimensions?
print(len(df_norm))
for count in range(len(df_norm)):
    xs = df_norm[count]['air_pressure'] #air_pressure
    ys = df_norm[count]['air_temp'] #air_temp
    zs = df_norm[count]['avg_wind_direction'] #avg_wind_direction
#   print("x: ",xs,"y: ",ys,"z: ",zs)
    ax.scatter(xs, ys, zs,alpha=0.5)
    count+=1
```
